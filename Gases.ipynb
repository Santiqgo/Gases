Realice diagramas Pv de isotermas para gases que obedecen las siguientes ecuaciones de estado
(realice los diagramas de las isotermas para distintos valores de a y b y compárelas):

(a) $Pv = RT$, ecuación del gas ideal.

(b) $(P + \frac{a}{v^{2}})(v − b) = RT$, ecuación de van der Waals.

(c) $P = \frac{RT}{v−b} − \frac{a}{v(v−b)\sqrt{t}}$, ecuación de Redlich-Kwong.

(d)$Pe^{\frac{a}{RTv}}(v-b)=RT$, ecuación de Dieterici

(e)$(P+\frac{a}{v^{2}T})(v-b)=RT$,ecuación de Berthelot

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
a=[0.2,0.011,0.34,0.130]
b=[0.02,0.10,0.11,0.200]
R=8.314472 #J/K*mol
T=293.15#K
V=np.linspace(10**-3,0.25,100)#m^3
```

```python
def ideal(x):
    return R*T/x
Pideal=ideal(V)*(10**-6)

plt.plot(V,Pideal)
plt.grid()
plt.title("Gas ideal")
plt.xlabel("Volumen (m)$^{2}$")
plt.ylabel("Presion (KPa)")
plt.show()
```

```python
def van(x,y,z):
    return ((R*T)/(x-z)) -(y/((x)**2))

Pvan1=van(V,a[0],b[0])*(10**-6)
Pvan2=van(V,a[1],b[1])*(10**-6)
Pvan3=van(V,a[2],b[2])*(10**-6)
Pvan4=van(V,a[3],b[3])*(10**-6)

plt.plot(V,Pvan1,"--")
plt.plot(V,Pvan2,"--")
plt.plot(V,Pvan3,"--")
plt.plot(V,Pvan4,"--")
plt.grid()
plt.title("Van der Waals")
plt.xlabel("Volumen (m)$^{2}$")
plt.ylabel("Presion (KPa)")
plt.show()
```

```python
def RK(x,y,z):
    return ((R*T)/(x-z)) - (y/(x*(x-z)*T**(0.5)))

PR1=RK(V,a[0],b[0])*(10**-6)
PR2=RK(V,a[1],b[1])*(10**-6)
PR3=RK(V,a[2],b[2])*(10**-6)
PR4=RK(V,a[3],b[3])*(10**-6)

plt.plot(V,PR1,"--")
plt.plot(V,PR2,"--")
plt.plot(V,PR3,"--")
plt.plot(V,PR4,"--")
plt.grid()
plt.title("Redlich-Kwong")
plt.xlabel("Volumen (m)$^{2}$")
plt.ylabel("Presion (KPa)")
plt.show()
```

```python
def D(x,y,z):
    return ((R*T)/(x-z))*np.exp(-y/(R*T*x))

PD1=D(V,a[0],b[0])*(10**-6)
PD2=D(V,a[1],b[1])*(10**-6)
PD3=D(V,a[2],b[2])*(10**-6)
PD4=D(V,a[3],b[3])*(10**-6)

plt.plot(V,PD1,"--")
plt.plot(V,PD2,"--")
plt.plot(V,PD3,"--")
plt.plot(V,PD4,"--")
plt.grid()
plt.title("Dieterici")
plt.xlabel("Volumen (m)$^{2}$")
plt.ylabel("Presion (KPa)")
plt.show()
```

```python
def B(x,y,z):
    return (R*T)/(x-z)-(y/(T*(x**2)))

PB1=B(V,a[0],b[0])*(10**-6)
PB2=B(V,a[1],b[1])*(10**-6)
PB3=B(V,a[2],b[2])*(10**-6)
PB4=B(V,a[3],b[3])*(10**-6)

plt.plot(V,PB1,"--")
plt.plot(V,PB2,"--")
plt.plot(V,PB3,"--")
plt.plot(V,PB4,"--")
plt.grid()
plt.title("Berthelot")
plt.xlabel("Volumen (m)$^{2}$")
plt.ylabel("Presion (KPa)")
plt.show()
```

Realice un estudio (grafique e interprete) de la función de distribución de velocidades de
Maxwell-Boltzmann, f(v), para distintos valores (cuatro como mínimo) de la masa m de las
partículas y de la temperatura, T. La función está dada por:

$f(v)=4\pi(\frac{m}{2\pi KT})^{\frac{3}{2}}v^{2}e^{-\frac{mv^{2}}{2KT}}$

Donde v es la velocidad de la partícula y k es la constante de Boltzmann ($1.381*10^{-23}\frac{J}{K}$.).

```python
K=1.381*10**-23 #J/K
T1=[273.15,284.16,150.11,326.53]
m=[1*10**-26,4*10**-26,6*10**-26,10*10**-26]
v=np.linspace(0,2580,4000)
```

```python
def f(x,y,z):
    return 4*np.pi*((z/(2*np.pi*K*y))**(3/2))*(x**2)*np.exp(-(z*x**2)/(2*K*y))
```

```python
F1=f(v,T1[0],m[0])
F2=f(v,T1[1],m[1])
F3=f(v,T1[2],m[2])
F4=f(v,T1[3],m[3])

plt.plot(v,F1,"--")
plt.plot(v,F2,"--")
plt.plot(v,F3,"--")
plt.plot(v,F4,"--")
plt.grid()
plt.title("Probabilidad de Maxwell-Boltzmann")
plt.xlabel("Velocidad (m/s)")
plt.ylabel("f(v)")
plt.show()
```
